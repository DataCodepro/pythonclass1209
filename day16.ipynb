```python
#1 Write a query to list the number of jobs available in the employees table
mycursor.execute('select distinct JobRole from employeedata')
for x in mycursor:
    print(x)
```

```python
#2 Write a query to get the total monthly income payable to employees
mycursor.execute('select sum(MonthlyIncome)  from employeedata')
for x in mycursor:
    print(x)
```

```python
#3.Write a query to get the minimum monthly income from employees table
mycursor.execute('select min(MonthlyIncome)  from employeedata')
for x in mycursor:
    print(x)
```

```python
#4.Write a query to get the maximum monthly income of an employee working as a Research Scientist
mycursor.execute('select max(MonthlyIncome)  from employeedata where JobRole = "Research Scientist"')
for x in mycursor:
    print(x)
```

```python
#5.Write a query to get the average monthly income and number of employees working as a Laboratory Technician
mycursor.execute('select avg(MonthlyIncome)  from employeedata where JobRole = "Laboratory Technician"')
for x in mycursor:
    print(x)
```

```python
#6 Write a query to get the number of employees with the same job
mycursor.execute('select JobRole, Count(JobRole)  from employeedata group by JobRole')
for x in mycursor:
    print(x)
```

```python
#7.Write a query to get the difference between the highest and lowest  monthly income
mycursor.execute('select max(MonthlyIncome) -  min(MonthlyIncome) from employeedata')
for x in mycursor:
    print(x)
```

```python
#8.Write a query to display the department and salary for all employees whose monthly income is not in the range 5000 through 7500
mycursor.execute('select Department, MonthlyIncome from employeedata where MonthlyIncome not between 5000 and 7500')
for x in mycursor:
    print(x)
```

```python
#9.Write a query to display the departmentand job of employees whose jobname have exactly 7 characters
mycursor.execute('select Department, JobRole from employeedata where JobRole like "_______"')
for x in mycursor:
    print(x)
```

```python
#10 Write a query to display the job of employees having 's' as the third character. 
mycursor.execute('select JobRole from employeedata where JobRole like "__s%"')
for x in mycursor:
    print(x)
```
